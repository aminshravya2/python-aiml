```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
df=pd.read_csv("C:\\Users\\GPT-BANTWAL\\Downloads\\mtcars .csv")
df
```

```python
range=(10,40)
bins=10
plt.hist(df['mpg'],bins,range,color="green",rwidth=0.9)
plt.xlabel('mpg')
plt.ylabel('frequency')
plt.title("histogram")
plt.show()
```

```python
plt.scatter(df['wt'],df['mpg'],color="black",marker="*")
plt.xlabel("weight of the car")
plt.ylabel("mpg")
plt.title("sactter plot")
plt.show()
```
